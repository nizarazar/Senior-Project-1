## This first section will go over proving BP3. The equation goes as follows: - 
$$
\frac{\partial C}{\partial b_j^l} = \delta_j^l
$$
Using BP2: - 
$$
\delta^l = ((W^{l+1})^T\delta^{l+1}) \odot \sigma'(z^l)
$$

### Here, we will start the step by step drivation of it. 
Since we know that $z^l$ is defined as:

$$
z^l = W^l a^{l-1} + b^l
$$

We can start by taking the derivative with respect to $b^l$:

$$
\frac{\partial z^l}{\partial b^l} = 1
$$

Now we can apply the chain rule, which gives us:

$$
\frac{\partial C}{\partial b^l} = \frac{\partial C}{\partial z^l} \cdot \frac{\partial z^l}{\partial b^l}
$$

Since $\frac{\partial z^l}{\partial b^l} = 1$, then we get:

$$
\frac{\partial C}{\partial b_j^l} = \delta_j^l
$$

Now, we are done proving BP3.


## The section will go over proving BP4. The equation goes as follows:-
$$
\frac{\partial C}{\partial W_{jk}^l} = a_k^{l-1} \delta_j^l
$$

### Here, we will start the step by step drivation of it. 
Using the known forward propagation equation, which goes as follows:

$$
z_j^l = \sum_k W_{jk}^l a_k^{l-1} + b_j^l
$$

We can take the derivative with respect to $W_{jk}^l$:

$$
\frac{\partial z_j^l}{\partial W_{jk}^l} = a_k^{l-1}
$$

Again we can use the chain rule:

$$
\frac{\partial C}{\partial W_{jk}^l} = \frac{\partial C}{\partial z_j^l} \cdot \frac{\partial z_j^l}{\partial W_{jk}^l}
$$

Now, $\delta_j^l = \frac{\partial C}{\partial z_j^l}$, so we get:

$$
\frac{\partial C}{\partial W_{jk}^l} = a_k^{l-1} \delta_j^l
$$

Done with proving BP4.
